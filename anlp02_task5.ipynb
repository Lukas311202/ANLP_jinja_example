## the inputs

```python
legal_letters = [
  {
    "id": 1,
    "letter_text": "To the Administrator of 'MuskGamingFails.com': We represent Mr. Elon Musk. It has come to our attention that your website hosts and indexes a curated repository of video clips displaying our client engaging in video game streams with disparaging commentary. This unauthorized aggregation infringes upon our client's publicity rights and constitutes ongoing harassment. We hereby demand that you permanently deactivate the domain and remove all associated content from the internet immediately. Failure to comply will result in immediate legal action.",
    "expected_output": {
      "content": "Demand to shut down a website tracking and hosting clips of Elon Musk playing video games poorly.",
      "sender": "Legal representatives of Elon Musk",
      "type": "cease and desist",
      "demands_money": False,
      "amount": None,
      "deadline": "immediate"
    }
  },
  {
    "id": 2,
    "letter_text": "Dear Student, Our records indicate that you have recently commenced part-time employment. Under statutory regulations, this change in income means you no longer qualify for complimentary coverage under your parents' family health insurance plan. To maintain continuous healthcare coverage, you are required to register for an individual student health insurance policy. Please complete and submit the enclosed registration forms to our offices no later than 15.08.2026.",
    "expected_output": {
      "content": "Notification that a student is no longer eligible for family health insurance due to a part-time job and must register individually.",
      "sender": "Health Insurance Provider",
      "type": "insurance notification",
      "demands_money": False,
      "amount": None,
      "deadline": "15.08.2026"
    }
  },
  {
    "id": 3,
    "letter_text": "FINAL NOTICE OF OUTSTANDING DEBT. This office has been retained by Nintendo of Europe to collect a delinquent balance on your account. Despite prior invoices, your payment for digital marketplace purchases remains outstanding. You are hereby ordered to remit the total amount of 69,99 Euros to the inkasso account listed below. Settling this balance by 21.06.2026 will halt further escalation, interest accrual, and potential legal proceedings.",
    "expected_output": {
      "content": "Debt collection demand for an unpaid balance owed to Nintendo for digital purchases.",
      "sender": "Inkasso (Debt Collection Agency)",
      "type": "payment demand",
      "demands_money": True,
      "amount": "69,99 Euros",
      "deadline": "21.06.2026"
    }
  }
]

legal_types = ["payment demand", "insurance notification", "cease and desist"]
essential_properties = [
  {"key": "content", "description": "Brief summary of the letters content"},
  {"key": "type", "description": "The type of the letter"},
  {"key": "deadline", "description": "The Deadline to fullfill any demanded obligation"},
  {"key": "sender", "description": "Sender of the letter"},
]
```

```python
from pathlib import Path
import json
from collections import Counter

PROJECT_DIR = Path("anlp_02_task5")
DATA_DIR = PROJECT_DIR / "data"
TEMPLATES_DIR = PROJECT_DIR / "templates"
OUTPUTS_DIR = PROJECT_DIR / "outputs"

for directory in [DATA_DIR, TEMPLATES_DIR, OUTPUTS_DIR]:
    directory.mkdir(parents=True, exist_ok=True)

PROJECT_DIR.resolve()
```

## The rendering

```python
from jinja2 import Environment, FileSystemLoader, StrictUndefined

env = Environment(
    loader=FileSystemLoader(PROJECT_DIR),
    undefined=StrictUndefined,
    trim_blocks=True,
    lstrip_blocks=True,
)

prompt_template = env.get_template("input_template.txt.j2")

for letter in legal_letters:
    rendered_prompt = prompt_template.render(
        letter=letter,
        legal_types=legal_types,
        essential_properties=essential_properties
    )
    (OUTPUTS_DIR / f"letter_{letter["id"]}.txt").write_text(rendered_prompt, encoding="utf-8")
```
